```python
X = heart_data.drop(columns='target', axis=1)
Y = heart_data['target']
```

```python
print(X)
```

```python
print(Y)
```

```python
X_train, X_test, Y_train, Y_test =train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
model = LogisticRegression()
```

```python
# training the LogisticRegression model with Training data
model.fit(X_train, Y_train)
```

```python
#accuracy on training data
X_train_pridiction = model.predict(X_train)
traning_data_accuracy = accuracy_score(X_train_pridiction,Y_train)
```

```python
print('accuracy on training data : ',traning_data_accuracy)
```

```python
#accuracy on test data
X_test_pridiction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_pridiction,Y_test)
```

```python
print('accuracy on test data : ',test_data_accuracy)
```
